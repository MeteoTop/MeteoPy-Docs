# 测试MetroDraw文件夹下的Draw_Nomap.py

## 测试Draw_Nomap类

```python
# # 安装MeteoPy之后，正式使用不需要这两句 # #
import sys
sys.path.append('../../')
# # # # # # # # # # # # # # # # # # # # # #
import numpy as np
import pandas as pd
from MeteoPy import Draw_Nomap
import matplotlib.pyplot as plt
```

```python
# # 读取测试数据
data = pd.read_excel('../TestData/Qs_Ts.xlsx', index_col=0)
x, y = data.iloc[:, 0], data.iloc[:, 1]
```

```python
# # 测试
# 创建画布和绘图对象
fig = plt.figure(figsize=(2.6, 1.8), dpi=200)
ax = fig.add_subplot(1, 1, 1)
# 绘图
draw = Draw_Nomap()
draw.format_setting(ax, x_axis={'xticks':np.arange(0, 50, 10), 'xlabel':r'$Q_s(MJ·m^{-2})$'}, 
                    y_axis={'yticks':np.arange(-4, 8, 2), 'ylabel':r'$\vartriangle T_s$'}, 
                    style='brief', tags={'(a)':[0.08, 0.95, {'font':{'size':12}}], 
                                         'AC':[0.5, 1.05, {'font':{'color':'blue'}}]})
draw.draw_scatter(fig, ax, x, y, density=True, 
                  fit_line={'tag':{'position':[0.2, 0.2], 'font':{'size':10}}},
                  type={'cbar':{'font_tick':{'color':'red', 'size':10}, 
                                'yticks':np.arange(0.002, 0.014, 0.002), 
                                'label':'Scatter Density', 
                                'font_label':{'size':10}}})
```
